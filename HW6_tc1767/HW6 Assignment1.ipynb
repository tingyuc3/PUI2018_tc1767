```python
mngrouped = dfM.groupby(['LandUse'],axis = 0).count()
mngrouped = mngrouped.reset_index()
mngrouped
```

```python
bkgrouped = dfB.groupby(['LandUse'],axis = 0).count()
bkgrouped = bkgrouped.reset_index()
bkgrouped
```

```python
bkgrouped = bkgrouped.astype(int)
bkgrouped.head()
```

```python
mngrouped = mngrouped.astype(int)
mngrouped.head()
```

```python
both = bkgrouped.merge(mngrouped, on = 'LandUse')
both.columns
```

```python
both = both.rename(columns = {"ZipCode_x": "bk", "ZipCode_y": "mn"})
both
```

```python
both['LandUse'][0] = 'One &Two Family Buildings'
both['LandUse'][1] = 'Multi-Family Walk-Up Buildings'
both['LandUse'][2] = 'Multi-Family Elevator Buildings'
both['LandUse'][3] = 'Mixed Residential & Commercial Buildings'
both['LandUse'][4] = 'Commercial & Office Buildings'
both['LandUse'][5] = 'Industrial & Manufacturing'
both['LandUse'][6] = 'Transportation & Utility'
both['LandUse'][7] = 'Public Facilities & Institutions'
both['LandUse'][8] = 'Open Space & Outdoor Recreation'
both['LandUse'][9] = 'Parking Facilities'
both['LandUse'][10] = 'Vacant Land'
both
```

```python
both['luratio'] = both.mn / both.bk
both
```

```python
leng = len(both)
def change(x):
    if x > 1:
        return'True'
    else:
        return 'False'
both['Manhattan Wins'] = both['luratio'].apply(change)
both
```

## comment on what you found out. Does it make sense?¶

Manhattan only has more Multi-Family Elevator Buildings than Brooklyn does. However, for all other land use classes, Brooklyn has more buildings than Manhattan does. It does make sense because Brooklyn is much lareger than Manhattan. Since Manhattan is so compact, it's reasonable that it has more multi-family elevator building in order to satisfy the need of large population living in Manhattan. On the other hand, since Brooklyn has larger acreage and lower living cost, there are more multi-family walk-up buildings and one&two family buildings.
