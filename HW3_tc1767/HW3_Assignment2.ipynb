```python
url = "https://data.cityofnewyork.us/resource/x7bj-bybh.json"
```

```python
ca = pd.read_json(url)
```

```python
ca.head(7)
```

```python
cel_3c = ca[['pre_filing_date','initial_cost','total_est_fee']]
```

```python
cel_3c.rename(columns = {'pre_filing_date':'Pre- Filing Date','initial_cost':'Initial Cost','total_est_fee':'Total Est. Fee'},inplace = True)
```

```python
ax = cel_3c.plot(x='Pre- Filing Date', y='Total Est. Fee');
ax.set_ylabel("Number");
```

```python
cel_3c["Pre- Filing Date"] = cel_3c["Pre- Filing Date"].apply(lambda x: pd.Timestamp(x))
```

```python
ax = cel_3c.plot(x='Pre- Filing Date', y='Total Est. Fee');
ax.set_ylabel("Number");
```

Figure2: the estimate fee of working on an Antenna as a function of time between 2011 and 2017. There are few extreme number in 2013 but most total fees are under 1500 (in thousands).

```python
ax = cel_3c.plot(x='Pre- Filing Date', y='Initial Cost');
ax.set_ylabel("Number");
```

Figure3: the initial cost of working on an Antenna as a function of time between 2011 and 2017. There are few extreme number in 2013 but most total fees are under 100000.
